1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
plt.figure(figsize=(10,6))
palette = sns.color_palette("hls", 6)

sns.lineplot(
    x="month_number",
    y="total_profit",
    data=df,
    ci = None,
    palette=palette
)
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="month_number",
    y="total_profit",
    data=df,
    palette=palette
)
plt.show()
```
